### Null hypothesis: the gender of a member of the 19th german parliament ("Bundestag") can be modelled as a Bernoulli process with equal propability for female and male members

```python
from scipy.special import binom
import matplotlib.pyplot as plt
```

```python
p_male = 0.5
p_female = 1 - p_male
```

```python
size = 709
male = 490
female = size - male
```

```python
def binom_prob(n, k, p):
    assert p > 0
    assert p <= 1
    assert k <= n
    return  binom(n, k) * p**(k) * (1 - p)**(n - k)
```

```python
probs = []
for i in range(size):
    probs.append(binom_prob(size, i, p_male))
```

```python
plt.plot(range(size), probs)
plt.axvline(x = female, label = "Number of female members", color = "r")
plt.legend()
plt.title("Propability Distribution")
plt.xlabel("female members")
plt.ylabel("propability")
fig = plt.gcf()
fig.set_size_inches(10, 5)
plt.show()
```

```python
cumulative_probs = []
for i in range(len(probs)):
    cumulative_probs.append(sum([x for x in probs[0:i]]))
```

```python
plt.clf()
plt.plot(range(size), cumulative_probs)
plt.axvline(x = female, label = "Number of female members", color = "r")
plt.legend()
plt.title("Cumulative Propability Distribution")
plt.xlabel("female members")
plt.ylabel("cumulative propability")
fig = plt.gcf()
fig.set_size_inches(10, 5)
plt.show()
```

### Under the assumption that the gender of the members of the german parliament ("Bundestag") can be modelled as a Bernoulli process with equal propability for female and male members, the propability of having a parliament with the same or less female members than the 19th parliament (219 female out of 790) is:

```python
print(cumulative_probs[female])
```

### Therefore the null hypothesis can be rejected.
