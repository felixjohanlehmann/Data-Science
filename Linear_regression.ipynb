# Linear regression (simple example)

```python
# https://machinelearningmastery.com/solve-linear-regression-using-linear-algebra/
```

```python
from numpy import array
import matplotlib.pyplot as plt
from numpy.linalg import inv
```

```python
data = array([
[0.05, 0.12],
[0.18, 0.22],
[0.31, 0.35],
[0.42, 0.38],
[0.5, 0.49],
])
print(data)
X, y = data[:,0], data[:,1]
X = X.reshape((len(X), 1))
# plot dataset
plt.scatter(X, y)
plt.show()
```

```python
# linear least squares
b = inv(X.T.dot(X)).dot(X.T).dot(y)
print(b)
# predict using coefficients
yhat = X.dot(b)
# plot data and predictions
plt.scatter(X, y)
plt.plot(X, yhat, color='red')
plt.show()
```

```python
X = array([
[3, 4],
[1, 0],
[2, 4],
[4, -5],
[9, 12],
])

y = array([
[9],
[-1],
[5],
[2],
[17],
])
```

```python
# linear least squares
b = inv(X.T.dot(X)).dot(X.T).dot(y)

print(b)
```

```python
sum(b)
```

```python
# predict using coefficients
yhat = X.dot(b)
# plot data and predictions
print(yhat)
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X[:,0], X[:,1], y)

ax.scatter(X[:,0], X[:,1], yhat, color = "r")

ax.set_xlabel('x1 Label')
ax.set_ylabel('x2 Label')
ax.set_zlabel('y Label')

plt.show()
```

```python
fig = plt.figure()
plt3d = plt.figure().gca(projection='3d')

plt3d.plot_surface(X[:,0], X[:,1], yhat, alpha=0.2)

plt.show()
```

```python
fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X[:,0], X[:,1], yhat, alpha=0.2)

plt.show()
```
